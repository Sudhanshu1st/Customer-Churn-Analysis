## 1.4 Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

### 1.4.1 Churn Channel

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

Looks like around 10% people have Churned.

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

### 1.4.2 Sales channel

```python
channel = client_df[['id', 'channel_sales', 'churn']]
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count().unstack(level=1).fillna(0)
channel_churn = (channel.div(channel.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)
```

```python
plot_stacked_bars(channel_churn, 'Sales channel', rot_=30)
```

The churning customers have been distributed by 5 channels and MISSING Values have 7.6 % churn rate.

### Verify the hypothesis of price sensitivity being correlated with churn:

#### Merging the client_df and price_df DataFrames on the "id" column:

```python
merged_data = pd.merge(client_df, price_df, on="id")
```

```python
merged_data["price_sensitivity"] = merged_data["price_peak_var"] - merged_data["price_off_peak_var"]
```

```python
sns.boxplot(x="churn", y="price_sensitivity", data=merged_data)
plt.show()
```

### Summary



Based on the exploratory data analysis (EDA) conducted on the provided data, the key findings are as follows:

1. Notable Trends:
   - There is a wide range of electricity and gas consumption among customers, with some having significantly higher or lower consumption compared to others.
   - The number of active products and services varies across customers, suggesting different levels of engagement and utilization.
   - There is a variation in the subscribed power (pow_max) among customers, indicating different energy needs.

2. Distributions:
   - The distributions of variables such as cons_12m, cons_gas_12m, forecast_cons_12m, and forecast_cons_year are skewed to the right, indicating that a majority of customers have lower consumption values, while a small portion has higher consumption levels.
   - The net_margin variable also exhibits a skewed distribution, suggesting varying profitability levels among customers.
   - The churn variable is binary (1 or 0), representing customers who churned and those who did not.

3. Correlations:
   - There are some notable correlations between certain variables. For example, there is a positive correlation between cons_12m and cons_last_month, indicating that customers with higher historical consumption tend to have higher recent consumption as well.
   - Additionally, there is a positive correlation between margin_gross_pow_ele and margin_net_pow_ele, suggesting a relationship between gross and net margins on power subscription.

4. Relationship between Price Sensitivity and Churn:
   - By merging the client_data and price_data, a measure of price sensitivity (price_sensitivity) was calculated as the difference between the price of energy during peak and off-peak periods.
   - A box plot analysis comparing price sensitivity with churn revealed that customers who churned tend to exhibit higher price sensitivity. This suggests that price changes may influence churn behavior, supporting the hypothesis that price sensitivity is correlated with churn to some extent.
